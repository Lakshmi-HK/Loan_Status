```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
url = "https://raw.githubusercontent.com/lemoinef/Loan-Default-Prediction/master/loans_2007.csv"
data = pd.read_csv(url)
```

```python
# Dataset_information
print('Rows and Columns:',  data.shape)
print('Data Type:', data.dtypes)
```

```python
print(data.isnull().sum())

```

```python
data = data.dropna()
print(data.isnull().sum())
```

```python
data.columns
```

```python
# drop employees id and similar infere column
data = data.drop(['id', 'member_id', 'funded_amnt', 'funded_amnt_inv', 'sub_grade', 'title'], axis = 1 )
```

```python
categorical_data = data.select_dtypes(exclude=["number","bool_"])
print('Categorical_data:',categorical_data.shape)
numerical_data = data.select_dtypes(exclude=["object"])
print('Numerical_data:', numerical_data.shape)
```

```python
categorical_data['loan_status'].value_counts()
mapping_dict = {
    "loan_status": {
        "Fully Paid": "Fully Paid",
        "Charged Off": "Charged Off",
        "Does not meet the credit policy. Status:Fully Paid": "Fully Paid",
        'Does not meet the credit policy. Status:Charged Off': "Charged Off",
        'Current': 'Fully Paid',
        'Late (31-120 days)': "Charged Off",
        'In Grace Period':'Fully Paid',
        'Late (16-30 days)' : 'Charged Off',
        'Default': 'Charged Off'
        }
}

categorical_data = categorical_data.replace(mapping_dict)

print(categorical_data['loan_status'].value_counts())
status = categorical_data['loan_status'].value_counts()
mylabels = ["Fully Paid", "Charged Off"]
plt.pie(status, labels = mylabels)
plt.savefig('loan_status')
plt.show() 
```

Fully Paid loans are more incomparison to charged off loans.

```python
print(categorical_data['term'].value_counts())
terms = categorical_data['term'].value_counts()
mylabels = ["36 months", "60 months"]
plt.pie(terms, labels = mylabels)
plt.savefig('term')
plt.show() 
```

36 months duration's loan perfered by most of the peoples.

```python
print(categorical_data['grade'].value_counts())
grade = categorical_data['grade'].value_counts()
mylabels = ["B", "A", "C", "D", "E", "F", "G"]
plt.pie(grade, labels = mylabels)
plt.show()
```

All grades of peoples pereferd to take loan, but here specially upper grade (A,B and C) peoples take loan more incomparison to lower grades peoples

```python
print(categorical_data['home_ownership'].value_counts())
home = categorical_data['home_ownership'].value_counts()
mylabels = ["Rent", "Mortage", "Own", "Other"]
plt.pie(home, labels = mylabels)
plt.show()
```

Mostly peoples prefered rent based accomodation and mortage their home for loans

```python
mapping_dict = {
    "verification_status": {
        "Not Verified": "Not Verified",
        "Verified": "Verified",
        "Source Verified": "Verified",
        }
}

categorical_data = categorical_data.replace(mapping_dict)
print(categorical_data['verification_status'].value_counts())
verification = categorical_data['verification_status'].value_counts()
mylabels = ["Not Verified", "Verified"]
plt.pie(verification, labels = mylabels)
plt.show()
```

Non verified peoples are more incomparison to varified peoples. 

```python
mapping_dict = {
    "emp_length": {
        "10+ years": 10,
        "9 years": 9,
        "8 years": 8,
        "7 years": 7,
        "6 years": 6,
        "5 years": 5,
        "4 years": 4,
        "3 years": 3,
        "2 years": 2,
        "1 year": 1,
        "< 1 year": 0,
        "n/a": 0
    }
}
categorical_data = categorical_data.replace(mapping_dict)
plt.figure(figsize=(15,10))
ax = sns.countplot(x="emp_length",data=categorical_data,
              order = categorical_data["emp_length"].value_counts().index)
plt.title("Vertical Bar Plot of the Employee's tenure")
plt.xticks(rotation=270)
plt.show()
```

10 years and above peoples are taking loans more incomparison to lower experience peoples

```python
plt.figure(figsize=(15,10))
ax = sns.countplot(x="purpose",data=categorical_data,
              order = categorical_data["purpose"].value_counts().index)
plt.title("Vertical Bar Plot of the loan's purpose")
plt.xticks(rotation=270)
plt.show()
```

Mostly peoples are taking loans to payoff previous loans
